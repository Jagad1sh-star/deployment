```python
μ3 =measure of skewness =  ∑(x-x_bar)^3/n
μ4 =Measure of kurtosis =  ∑(x-x_bar)^4/n  

skew = μ3^2/μ2^3
kurt = μ4/ μ2^2
```

```python
μ2 = (hieght-hieght.mean())**2
μ3 = (hieght-hieght.mean())**3
μ4 = (hieght-hieght.mean())**4
```

```python
hight_data = pd.DataFrame({'μ1':hieght,
             'μ2':μ2,
             'μ3':μ3,
             'μ4':μ4})           
```

```python
skew = (hight_data['μ3'].mean())**2 / (hight_data['μ2'].mean())**3
skew
```

```python
hieght.skew()
```

```python
kurt = (hight_data['μ4'].mean()) / (hight_data['μ2'].mean())**2 # kurt == 3 
kurt
```

# Bar chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Rank = salaries.groupby(['rank','sex'])[['salary','phd']].mean().sort_values(by = 'salary',ascending = False)
```

```python
Rank
```

```python
Rank.plot(kind = 'bar',xlabel = 'Rank of professor',ylabel = 'Average salary for Professor',
          title = 'Average salary of professors by the Rank',color = ['r','g'],subplots = True,alpha = 0.5,table = True)
```

```python
Gender = salaries.groupby(['sex'])['salary'].mean().sort_values(ascending = False)
```

```python
Gender.plot(kind = 'bar',xlabel = 'Gender of professor',ylabel = 'Average salary for Professor',
          title = 'Average salary of professors by the Gender',color = 'r')
```
